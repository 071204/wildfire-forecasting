```python
import xarray as xr
```

```python
data = xr.load_dataset('LAI_mean_monthly_1981-2015.nc4')
```

```python
data
```

### Range (long, lat): `-21.75, 72.25` *top-left* to  `45.25, 26.75` *bottom-right*

```python
sub_data = data.LAI.sel(lon=slice(-21.75, 45.25), lat=slice(26.75, 72.25))
```

```python
sub_data
```

```python
sub_data.values
```

```python
sub_data[0].values.max()
```

```python
sub_data[0].plot()
```

```python
sub_data = sub_data.sortby('lat', ascending=False)
```

```python
sub_data
```

```python
sub_data.to_netcdf(path='../data/stage.nc4', format='NETCDF4')
```
